```python
import matplotlib.pyplot as plt
data = [5, 7, 7, 8, 10, 11, 30, 40, 30, 50, 14, 14]
plt.hist(data, bins=10, edgecolor='red')
plt.title('simple Histogram example')
plt.ylabel('count')
plt.xlabel('numbers')
plt.show()
```

```python
import  pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
datf = pd.DataFrame({"season 1": [7, 7, 5, 5, 6, ],
                     "Season 2" : [1, 2, 4, 5, 7,]})
p = sns.histplot(data = datf)
p.set(xlabel="X label value", ylabel = "y Label Value")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
data = np.random.randint(20, 71, 1000)
plt.hist(data, bins=15, edgecolor='black', color= "pink")
plt.title('simple Histogram  of cancer patients age distribution')
plt.ylabel('Age')
plt.xlabel('numbers of patients')
plt.show()
```
